```python
f = open("whis.txt","r")
#first = f.readline()
#t = first.split("	")
#print(t)
whists = []
count = 0
for x in f:
    t = x.split("	")
    if count == 0:
        print(t)
    data = []
    
    data.append(t[2])
    data.append(t[5])
    data.append(count)
    whists.append(data)
    count += 1
    
for x in whists:
    print(x)
```
